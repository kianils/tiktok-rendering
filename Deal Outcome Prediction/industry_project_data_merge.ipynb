# load the mapping json file

### The mapping json file contains the dictionaries that can be used to map any of the 2 files together

```python
### read in the mapping file
import json

with open('mappings.json', 'r') as f:
    mappings = json.load(f)

### check the keys
print(f'mapping keys: {mappings.keys()}')
```

- `CompanyToDeals`: contains the mappings from company ID to Deals ID
- `CompanyToTickets`: contains the mappings from company ID to Tickets ID
- `TicketToDeal`: contains the mappings from Ticket ID to Deal ID

# Map company and deals dataset

### In order to map the company data and the deals data, we need to first create a one to one mapping between the deals ID and the company ID based on mappings['CompanyToDeals']

```python
# revert the company-deal mapping
deal_to_company = {}

for company_id, deal_list in mappings["CompanyToDeals"].items():
    for deal_id in deal_list:
        deal_to_company[deal_id] = company_id
deal_to_company
```

### Add a company ID column in the deals dataset for table merging

```python
### record ID is the unique deal ID
### first convert it into string type to match the datatype in the dictionary
deals['Record ID'] = deals['Record ID'].astype(str)

#### create a company for company id based on the dictionary using mapping function

deals['Company ID'] = deals['Record ID'].map(deal_to_company)
```

### Merge the deals dataset and the companies dataset

commpany id from the deals table: 'Company ID'

company id from the company table: 'Record ID'

```python
### convert the integer type to string type for Record ID
companies['Record ID'] = companies['Record ID'].astype(str)


deals_merged = pd.merge(left = deals, right = companies, left_on ='Company ID'  ,right_on ='Record ID' , how = 'left')
```
